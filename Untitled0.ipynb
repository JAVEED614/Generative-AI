```python
# üöÄ Install dependencies
!pip install -q transformers datasets accelerate

# üìÇ Upload your dataset (e.g., training_data.txt)
from google.colab import files
uploaded = files.upload()

# üßπ Load and tokenize the dataset
from datasets import Dataset, DatasetDict
from transformers import GPT2Tokenizer

file_name = list(uploaded.keys())[0]
file_content = uploaded[file_name].decode("utf-8")

# Create a list of text lines
lines = file_content.splitlines()

# Load the dataset from lines
dataset = Dataset.from_dict({"text": lines})

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token  # Prevents size mismatch error

def tokenize_function(example):
    return tokenizer(
        example["text"],
        padding="max_length",
        truncation=True,         # ‚úÖ Fix: explicitly enable truncation
        max_length=128,
    )

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Add labels for language modeling
tokenized_dataset = tokenized_dataset.map(lambda examples: {"labels": examples["input_ids"]}, batched=True)

# Create a DatasetDict
train_dataset = DatasetDict({"train": tokenized_dataset})

# üì• Load GPT-2 model
from transformers import GPT2LMHeadModel
model = GPT2LMHeadModel.from_pretrained("gpt2")
model.resize_token_embeddings(len(tokenizer))

# ‚öôÔ∏è Training Setup
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./gpt2-finetuned",
    per_device_train_batch_size=2,
    num_train_epochs=3,
    logging_steps=10,
    save_steps=100,
    save_total_limit=2,
    prediction_loss_only=True,
    report_to="none"  # disables WandB or other loggers
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset["train"],
)

# üîÅ Train the model
trainer.train()

# üí¨ Generate text
from transformers import pipeline

generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

prompt = "In the future, AI will"
result = generator(prompt, max_new_tokens=100, num_return_sequences=1)  # ‚úÖ Fix: use max_new_tokens instead of max_length
print(result[0]["generated_text"])
```
